```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("C:\\Users\\GPT BANTWAL\\Downloads\\mtcars (1).csv")
df
```

```python
range=(10,40)
bins=10
plt.hist(df['mpg'],bins,range,color='steelblue',rwidth=0.8)
plt.xlabel('mpg')
plt.ylabel('Frequency')
plt.title('my histogram')
plt.show()
```

```python
plt.scatter(df["wt"],df["mpg"],color="maroon",marker="*")
plt.xlabel("weight of the car")
plt.ylabel("mpg")
plt.title("scatter plot")
```

```python
temp=df['am'].value_counts()
temp.plot.bar(color="skyblue")
plt.title("Frequency distribution of transmission types")
plt.xlabel("Transmission types")
plt.ylabel("Count")
```

```python
plt.boxplot(df['mpg'],(color=="red"))
plt.ylabel('mpg',color="blue")
plt.title('boxplot',color="green")
```
