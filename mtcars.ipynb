```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
cars_df=pd.read_csv("C:\\Users\\CSINFO17\\Downloads\\mtcars (1).csv")
columns=["model","mpg","cyl","disp","hp","drat","wt","qsec","vs","am","gear","carb"]
cars_df[columns].head()
```

```python
range=(0,30)
bins=10
plt.hist('mpg',range,bins,color='steelblue',histtype='bar',width=0.8)
plt.xlabel('mpg')
plt.ylabel('Frequency')
plt.title('my histogram')
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(x='wt', y='mpg', data=cars_df)
plt.title('Scatter Plot of Car Weight vs MPG')
plt.xlabel('Weight')
plt.ylabel('MPG')
plt.show()
```

```python
temp= cars_df['am'].value_counts()
print(temp)
temp.plot.bar()
plt.title("Frequency distribution of transmission types")
plt.xlabel("Transmission types")
plt.ylabel("Count")
```

```python
plt.figure(figsize=(0, 1))
plt.box(cars_df[ 'mpg'])
plt.title('Box and Whisker Plot of MPG')
plt.ylabel('Miles Per Gallon (mpg)')
plt.show()
```
